# Data dictionary

## Data uit **intensiteit-snelheid**
| Kolomnaam | Beschrijving |
|------------|--------------|
| `id_meetlocatie` | Identiteit van de meetlocatie |
| `start_meetperiode` | Start van de periode waarover de gegevens in deze rij staan |
| `eind_meetperiode` | Einde van de periode waarover de gegevens in deze rij staan |
| `ndw_index` | Combinatie van voertuigcategorie en rijstrook |
| `gem_intensiteit` | Gemiddelde intensiteit (voertuigen/uur) |
| `gem_snelheid` | Gemiddelde snelheid (km/u) |
| `rijstrook_rijbaan` | Rijstrooknummer vanaf de middenberm |
| `totaal_aantal_rijstroken` | Totaal aantal rijstroken van de meetlocatie |
| `specifieke_baan` | Type rijbaan (mainCarriageway = hoofdrijbaan) |
| `locatiecode_vild` | Locatiecode van het VILD-punt |

---

## Data uit **file berichten**          - Moet nog toegevoegd worden aan pipeline
| Kolomnaam | Beschrijving |
|------------|--------------|
| `file_id` | Identiteit van de file |
| `file_starttijd` | Start van de periode waarover de gegevens over desbetreffende file gaan |
| `file_eindtijd` | Eind van de periode waarover de gegevens van desbetreffendez file gaan |
| `algemeen_oorzaak_type` | Bij een managed cause is er sprake van een verwachte/beheerde oorzaak; anders non-managed, onvoorspelbaar |
| `oorzaak_beschrijving` | Nadere toelichting op de oorzaak |
| `oorzaak_type` | Typering van de oorzaak (bijv. accident, earlierAccident, frost, etc.) |
| `vild_primaire_wegnummer` | Uniek nummer van de VILD-locatie van desbetreffende weg |
| `vild_primair_richting` | De richting waarop de file rijd |

---

## Data uit **geopanda**
| Kolomnaam | Beschrijving |
|------------|--------------|
| `wegtype` | De eerste letter van het type van de weg (hoofdrijbaan, oprit, afrit, etc.) |
| `geometry` | Coördinaten van de locatie |

---

## Data handmatig gegroepeerd
| Kolomnaam | Beschrijving |
|------------|--------------|
| `datum` | Datum van de meting |
| `tijd` | Tijdstip van de meting |
| `gem_snelheid_smooth` | gemiddelde snelheid smoothed met behulp van het [Savitzky–Golay filter](https://en.wikipedia.org/wiki/Savitzky%E2%80%93Golay_filter) |
| `gem_intensiteit_smooth` | gemiddelde intensiteit smoothed met behulp van het [Savitzky–Golay filter](https://en.wikipedia.org/wiki/Savitzky%E2%80%93Golay_filter) |

## Data KNMI

| Kolomnaam | Beschrijving |
|------------|--------------|
| `Tijd` | tijd (HH=uur; UT.12 UT=13 MET; 14 MEZT. Uurvak 05 loopt van 04.00 UT tot 5.00 UT |
| `Wind_snelheid` | Windsnelheid (in 0.1 m/s) gemiddeld over de laatste 10 minuten van het afgelopen uur |
| `Hoogste_wind_stoot` | Hoogste windstoot (in 0.1 m/s) over het afgelopen uurvak |
| `Temperatuur` | Temperatuur (in 0.1 graden Celsius) op 1.50 m hoogte tijdens de waarneming |
| `Min_temperatuur` | Minimumtemperatuur (in 0.1 graden Celsius) op 10 cm hoogte in de afgelopen 6 uur  |
| `Duur_neerslag` | Duur van de neerslag (in 0.1 uur) per uurvak |
| `Uursom_neerslag` | Uursom van de neerslag (in 0.1 mm) (-1 voor <0.05 mm) |
| `Horziontaal_zicht` | Horizontaal zicht tijdens de waarneming (0=minder dan 100m; 1=100-200m; 2=200-300m;...; 49=4900-5000m; 50=5-6km; 56=6-7km; 57=7-8km; ...; 79=29-30km; 80=30-35km; 81=35-40km;...; 89=meer dan 70km) |
| `Mist` | Mist 0=niet voorgekomen; 1=wel voorgekomen in het voorgaande uur en/of tijdens de waarneming |
| `Regen` | Regen 0=niet voorgekomen; 1=wel voorgekomen in het voorgaande uur en/of tijdens de waarneming |
| `Sneeuw` | Sneeuw 0=niet voorgekomen; 1=wel voorgekomen in het voorgaande uur en/of tijdens de waarneming |
| `Onweer` | Onweer 0=niet voorgekomen; 1=wel voorgekomen in het voorgaande uur en/of tijdens de waarneming |
| `IJsvorming` | IJsvorming 0=niet voorgekomen; 1=wel voorgekomen in het voorgaande uur en/of tijdens de waarneming |
